# Removendo nós isolados

Removendo os nós que ficaram isolados depois da tradução 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mygene
import h5py
import networkx as nx
import seaborn as sns
%matplotlib inline
```

## Rede Parsimonious Composite Network (PCNet)

```python
pcnet_network = pd.read_csv('C:/Users/renan/Desktop/UFRGS/GNN/data/networks/PCNET_network2.tsv', sep = '\t', index_col = 'Unnamed: 0')

pcnet_network
```

```python
comparison_column = np.where(pcnet_network["Source_Name"] == pcnet_network["Target_Name"], 'True', 'False')

pcnet_network["equal"] = comparison_column

pcnet_network
```

```python
pcnet_network['equal'].value_counts()
```

```python
df = pcnet_network[~pcnet_network['equal'].isin(['True'])]

df
```

```python
df['equal'].value_counts()
```

```python
df2 = df.drop(columns=['equal'])

df2.to_csv('C:/Users/renan/Desktop/UFRGS/GNN/data/networks/PCNET_network3.tsv', sep='\t')
```

```python
pcnet_network = pd.read_csv('C:/Users/renan/Desktop/UFRGS/GNN/data/networks/PCNET_network3.tsv', sep = '\t')

pcnet_network = pcnet_network.drop(columns=['Unnamed: 0'])

# Colocando cada linha em ordem alfabetica (Independente da coluna)

pcnet_network[['Source_Name','Target_Name']] = pd.DataFrame(np.sort(pcnet_network[['Source_Name','Target_Name']].values), index=pcnet_network.index, columns=['Source_Name','Target_Name'])

pcnet_network
```

```python
drop_duplicates = pcnet_network.drop_duplicates(subset=['Source_Name', 'Target_Name'], keep='last')

drop_duplicates
```

```python
drop_duplicates["ppi"] = drop_duplicates["Source_Name"] + drop_duplicates["Target_Name"]

drop_duplicates
```

```python
drop_duplicates_2 = drop_duplicates.drop_duplicates(subset=['ppi'], keep='last')

drop_duplicates_2
```

```python
drop_duplicates_3 = drop_duplicates_2.rename(columns={"Source_Name": "gene1", "Target_Name": "gene2"})

drop_duplicates_4 = drop_duplicates_3.drop(['ppi'], axis=1)

drop_duplicates_4
```

```python
drop_duplicates_4.to_csv('C:/Users/renan/Desktop/UFRGS/GNN/data/networks/PCNET_network3.tsv', sep='\t')
```
