### PCNET

```python
labels_pcnet = pd.read_csv('C:/Users/renan/Desktop/UFRGS/GNN/data/data_final/PCNET_labels.tsv', sep='\t')

labels_pcnet
```

```python
# Verifica quais genes do labels_unity (rede unificada, quantidade total de genes disponíveis) está na rede pcnet
# Após isso é só criar um nova coluna com a condição e remover os False
# A quantidade de True deve ser sempre a mesma de arestas da rede que estou tratando

labels_unity['gene'].isin(labels_pcnet['gene']).value_counts()
```

```python
labels_unity['temp'] = labels_unity['gene'].isin(labels_pcnet['gene'])

labels_unity
```

```python
labels_unity['temp'].value_counts()

# Conferir a quantidade da nova culuna, que deve ser a mesma do primeiro contador
```

```python
labels_pcnet2 = labels_unity[labels_unity['temp'] == True]

labels_pcnet2
```

```python
labels_pcnet3 = labels_pcnet2.drop(columns=['temp'])

labels_pcnet3
```

```python
labels_pcnet3['label'].value_counts(dropna = False)

# 12454 passengers (False)
# 5755 candidates (NaN)
# 907 drivers (True)
```

```python
labels_pcnet3.to_csv('C:/Users/renan/Desktop/UFRGS/GNN/data/data_final/PCNET_labels_semisupervised.tsv', sep='\t')
```
