# Tradução das redes

### PCNET

```python
# Leitura dos dicionários dos genes

pcnet_genes = pd.read_csv('C:/Users/renan/Desktop/UFRGS/GNN/data/traducao_redes/PCNET_lista_final.tsv', sep='\t', index_col='Unnamed: 0')

pcnet_genes
```

```python
pcnet_genes['Approved symbol'].isna().sum()
```

```python
# Procurar o caso dos genes duplicados e escritos diferentes

print(pcnet_genes.loc[pcnet_genes['Input'] == 'C8orf4'])
```

```python
pcnet_network
```

```python
print(pcnet_network.loc[pcnet_network['Source_Name'] == 'C8ORF4'])
```

```python
# Remover da rede quando existir o gene C19ORF12 com letras maiusculas (Só tem um caso a linha 988426) 
# porque ele não vai traduzir pelo problema do site que coloca essa entrada para letras minusculas, então esse tem que ser removido manualmente caso contrario não traduz e fica errado

# O mesmo acontece com o gene C8ORF4 com as letras maiuculas, no dicionario só tem a entrada em minusculas por causa do site, logo as ligações que tiverem estritas em maiusculo irão passar
# É o caso das linhas 1001504 e 1001505

# Removendo as linhas da rede: 

pcnet_network.drop([988426,1001504,1001505], axis=0, inplace=True)
```

```python
print(pcnet_network.loc[pcnet_network['Source_Name'] == 'C19ORF12'])
```

```python
# Tá dando problema de memória a operação de traduzir as duas maiores redes, como funcionou para a rede de um milhao, 
# vou dividir essa em duas e fazer a tradução das duas e depois concaternar em uma só de novo

k = 2
folds = np.array_split(pcnet_network, k)

temp0 = folds[0]
temp1 = folds[1]
```

```python
temp0.replace(pcnet_genes.set_index('Input').to_dict()['Approved symbol'], inplace = True)

temp0
```

```python
temp0['Target_Name'].isna().sum()
```

```python
temp1.replace(pcnet_genes.set_index('Input').to_dict()['Approved symbol'], inplace = True)

temp1
```

```python
temp1['Target_Name'].isna().sum()
```

```python
pcnet_network_trad2 = pd.concat([temp0, temp1])

pcnet_network_trad2
```

```python
pcnet_network_trad2['Target_Name'].isna().sum()
```

```python
PCNET2 = pcnet_network_trad2[pcnet_network_trad2.notna()]

PCNET2
```

```python
PCNET2['Target_Name'].isna().sum()
```

```python
PCNET3 = PCNET2.dropna()

PCNET3
```

```python
pcnet_network_trad = pd.merge(temp0, temp1, how = 'outer')

# Não estou mais usando o merge, usei o concat mais acima
```

```python
# Procurar um que eu sei que precisa ser traduzido C8ORF4 para TCIM

print(pcnet_network_trad.loc[pcnet_network_trad['Source_Name'] == 'C8ORF4'])
```

```python
pcnet_network_trad
```

```python
# Remover as ligações que tiverem NaN independente da coluna 

PCNET = pcnet_network_trad[pcnet_network_trad.notna()]

PCNET
```

```python
PCNET3.to_csv('C:/Users/renan/Desktop/UFRGS/GNN/data/networks/PCNET_network2.tsv', sep='\t')
```
