```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mygene
import h5py
import networkx as nx
import seaborn as sns
%matplotlib inline
```

```python
# a = crossentropy          = cross entropy 
# b = undersampling         = undersampling
# d = weighted_focalloss    = weighted focal loss (gamma=0, alpha=0.85) 
# e = focalloss_05          = focal loss (gamma=0.5, alpha=0.50) 
# f = focalloss_1           = focal loss (gamma=1, alpha=0.25) 
# g = focalloss_2           = focal loss (gamma=2, alpha=0.25) 

local = 'CPDB'
rede = 'cpdb'
paramset = 0

appended_data = []

# hprd_paramset0_fold1

# C:\Users\renan\Desktop\experiments\GCN\gridsearch/HPRD/prd_paramset0_fold1
```

```python

for paramset in range(288):

    fold1 = pd.read_csv(f'C:/Users/renan/Desktop/experiments/GCN/gridsearch/{local}/{rede}_paramset{paramset}_fold1.csv')
    fold1 = fold1.rename(columns={"Unnamed: 0": "epochs"})
    fold1['fold']=1

    fold2 = pd.read_csv(f'C:/Users/renan/Desktop/experiments/GCN/gridsearch/{local}/{rede}_paramset{paramset}_fold2.csv')
    fold2 = fold2.rename(columns={"Unnamed: 0": "epochs"})
    fold2['fold']=2

    fold3 = pd.read_csv(f'C:/Users/renan/Desktop/experiments/GCN/gridsearch/{local}/{rede}_paramset{paramset}_fold3.csv')
    fold3 = fold3.rename(columns={"Unnamed: 0": "epochs"})
    fold3['fold']=3

    fold4 = pd.read_csv(f'C:/Users/renan/Desktop/experiments/GCN/gridsearch/{local}/{rede}_paramset{paramset}_fold4.csv')
    fold4 = fold4.rename(columns={"Unnamed: 0": "epochs"})
    fold4['fold']=4

    fold5 = pd.read_csv(f'C:/Users/renan/Desktop/experiments/GCN/gridsearch/{local}/{rede}_paramset{paramset}_fold5.csv')
    fold5 = fold5.rename(columns={"Unnamed: 0": "epochs"})
    fold5['fold']=5

    # --------------------------------------------------------------
    
    paramset_concat = pd.concat([fold1, fold2, fold3, fold4, fold5])
    paramset_concat['paramset']=paramset

    appended_data.append(paramset_concat)
    
appended_data = pd.concat(appended_data) 
```

```python
appended_data.to_csv(f'C:/Users/renan/Desktop/experiments/GCN/gridsearch/{rede}_complete_gridsearch.tsv', sep='\t', index=False)
```
